```
import matplotlib.pyplot as plt
import numpy as np
```

```
%matplotlib inline
```

```
# Compute the x and y coordinates for points on a sine curve
x = np.arange(0, 3 * np.pi, 0.1)
y = np.sin(x)
```

```
print(x[:10], y[:10])
```

```
plt.plot(x, y)
```

```
plt.show()
```

```
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points using matplotlib
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

```
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
```

```
plt.subplot(1, 2, 1)
```

```
plt.plot(x, y_sin)
plt.title('Sine')
```

```
plt.subplot(1, 2, 2)
plt.plot(x, y_cos)
plt.title('Cosine')
```

```
plt.show()
```

```
import plotly.express as px
```

```
iris = px.data.iris()
print(iris.shape)
iris.head()
```

```
fig = px.scatter_3d(iris, 
              x='sepal_length', 
              y='sepal_width',
              z='petal_width',
              color='species')
fig.show()
```
